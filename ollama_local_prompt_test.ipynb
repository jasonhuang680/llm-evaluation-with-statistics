```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import PromptTemplate

print("Loading model, please wait 1-5 minutes...")

# Initialize the local Ollama model
llm = OllamaLLM(model="gpt-oss:20b")

# English prompt template (professional and clear)
template = """
You are an experienced statistics expert and data scientist.
Please answer the following question clearly, professionally, and concisely.
Use formal academic tone and provide examples when appropriate.

Question: {question}

Answer:
"""

prompt = PromptTemplate.from_template(template)

# Create the chain
chain = prompt | llm

# English question (same meaning as your original)
question = "Explain the meaning of p-value in hypothesis testing, and clarify why p < 0.05 is considered statistically significant but does not mean the effect is practically large. Provide a real-world example."

# Run the chain
response = chain.invoke({"question": question})

print("=== LLM Output ===")
print(response)
print("\n=== End ===")
```
